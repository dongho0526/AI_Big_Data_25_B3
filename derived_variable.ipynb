```python
# 데이터 전처리 
sales["구군"] = sales["구군"].apply(lambda x: x.strip())
sales['구군'] = sales['구군'].apply(lambda x: x.split()[0])
sales.loc[(sales['구군'] == '여주군'), '구군'] = '여주시'
sales['구군'] = sales['구군'].replace({'청원군': '청주시'})


incomes['구군'] = incomes['구군'].apply(lambda x: x.split()[0])
incomes.loc[(incomes['시'] == '세종자치시') & (incomes['구군'] == '소계'), '구군'] = '세종자치시'
```
